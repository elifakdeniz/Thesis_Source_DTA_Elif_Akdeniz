```python
import pandas as pd

# Load the datasets
df_syntactic = pd.read_csv("/Users/elifakdeniz/Desktop/Thesis_New/Notebooks/Jupyter_notebook/Future Engineering/syntatic features/final_dataset_with_syntactic_features.csv")
df_sentiment = pd.read_csv("/Users/elifakdeniz/Desktop/Thesis_New/Notebooks/Jupyter_notebook/Future Engineering/sentiment/2_final_dataset_deduplicated.csv")
df_tone = pd.read_csv("/Users/elifakdeniz/Desktop/Thesis_New/Notebooks/Jupyter_notebook/Future Engineering/job_postings_with_tone_features.csv")
```

```python
print(df_syntactic.columns)
print(df_sentiment.columns)
print(df_tone.columns)
```

```python
# Check shapes and preview
print("Syntactic shape:", df_syntactic.shape)
print("Sentiment shape:", df_sentiment.shape)
print("Tone shape:", df_tone.shape)

# Check missing values
print("Missing in syntactic:\n", df_syntactic.isnull().sum())
print("Missing in sentiment:\n", df_sentiment.isnull().sum())
print("Missing in tone:\n", df_tone.isnull().sum())

# Sample values
df_syntactic.sample(5)
df_sentiment.sample(5)
df_tone.sample(5)
```
