```python
import spacy
import re

nlp = spacy.load("en_core_web_sm")
```

```python
def spacy_clean_lemma(text):
    text = re.sub(r'(?i)job description[:\-]*', '', str(text))  # Remove job desc header
    doc = nlp(text)
    return [token.lemma_.lower() for token in doc if token.is_alpha and not token.is_stop]
from tqdm import tqdm
tqdm.pandas()

df['lemmas'] = df['description'].progress_apply(spacy_clean_lemma)
```

```python
df.to_csv("job_postings_with_lemmas.csv", index=False)
```
